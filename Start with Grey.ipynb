### Import necessary libraries and read CSV file

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as py

gdp = pd.read_csv('gapminder_gdp_oceania.csv')
```

### View data frame

```python
gdp.head()
```

### Transpose the DataFrame to get the years as rows instead of columns

```python
gdp.set_index('country', inplace=True)
gdp = gdp.T
gdp.index = [int(year.split('_')[1]) for year in gdp.index] 
```

### View data frame after changes

```python
gdp.head()
```

### Create plot starting with grey

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(gdp.index, gdp['Australia'], label='Australia', marker='o', color='lightgrey')
plt.plot(gdp.index, gdp['New Zealand'], label='New Zealand', marker='o', color='darkgrey')

# Add titles and labels (in grey)
plt.title('GDP Per Capita Comparison: Australia vs New Zealand (1952-2007)', color='grey')
plt.xlabel('Year', color='grey')
plt.ylabel('GDP Per Capita (USD)', color='grey')

# Customize the tick labels and legends in grey
plt.xticks(color='grey')
plt.yticks(color='grey')
plt.legend(facecolor='white', edgecolor='grey', labelcolor='grey')

# Show the grid (in grey)
plt.grid(True, color='grey', linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```

We should avoid legends when possible. How can we show the data?

### We decided to add country names to the end of the lines

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(gdp.index, gdp['Australia'], label='Australia', color='lightgrey', linewidth=3.5)
plt.plot(gdp.index, gdp['New Zealand'], label='New Zealand', color='darkgrey', linewidth=3.5)

# Add titles and labels (in grey)
plt.title('GDP Per Capita Comparison: Australia vs New Zealand (1952-2007)', color='grey')
plt.xlabel('Year', color='grey')
plt.ylabel('GDP Per Capita (USD)', color='grey')

# Customize the tick labels and legends in grey
plt.xticks(color='grey')
plt.yticks(color='grey')
#plt.legend(facecolor='white', edgecolor='grey', labelcolor='grey')

# Show the grid (in grey)
#plt.grid(True, color='grey', linestyle='--', linewidth=0.5)

# Add text labels at the end of each line
plt.text(gdp.index[-1], gdp['Australia'].iloc[-1], 'Australia', color='darkgrey', fontsize=12, va='bottom')
plt.text(gdp.index[-1], gdp['New Zealand'].iloc[-1], 'New Zealand', color='darkgrey', fontsize=12, va='bottom')

# Display the plot
plt.show()
```

### Clean up things by removing top and right borders (spines)

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(gdp.index, gdp['Australia'], label='Australia', color='lightgrey', linewidth=3.5)
plt.plot(gdp.index, gdp['New Zealand'], label='New Zealand', color='darkgrey', linewidth=3.5)

# Add titles and labels (in grey)
plt.title('GDP Per Capita Comparison: Australia vs New Zealand (1952-2007)', color='grey')
plt.xlabel('Year', color='grey')
plt.ylabel('GDP Per Capita (USD)', color='grey')

# Customize the tick labels and legends in grey
plt.xticks(color='grey')
plt.yticks(color='grey')
#plt.legend(facecolor='white', edgecolor='grey', labelcolor='grey')

# Show the grid (in grey)
#plt.grid(True, color='grey', linestyle='--', linewidth=0.5)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('grey')
ax.spines['left'].set_color('grey')

# Add text labels at the end of each line
plt.text(gdp.index[-1], gdp['Australia'].iloc[-1], 'Australia', color='darkgrey', fontsize=12, va='bottom')
plt.text(gdp.index[-1], gdp['New Zealand'].iloc[-1], 'New Zealand', color='darkgrey', fontsize=12, va='bottom')

# Display the plot
plt.show()
```
